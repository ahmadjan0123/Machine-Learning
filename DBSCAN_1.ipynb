```python
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
```

```python
x,y =  make_blobs(
    n_samples=300,
    cluster_std = 0.6,
    centers = 3,
    random_state=42
)
```

```python
plt.Figure(figsize=(8,6))
plt.scatter(
    x[: ,0],
    x[: ,1],
    cmap = 'virdis'

)
```

```python
model = DBSCAN(eps=0.5,min_samples = 4)
```

```python
model.fit_predict(x)
```

```python
labels = model.labels_
print(labels)
```

```python
plt.Figure(figsize=(8,6))

plt.scatter(
    x[:,0],
    x[:,1],
    c = labels,
    cmap='viridis'

)
plt.show()
```

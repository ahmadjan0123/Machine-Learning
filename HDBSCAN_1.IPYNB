```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import HDBSCAN
from sklearn.datasets import make_blobs
```

```python
x,y = make_blobs(
    n_samples=300,
    centers = 3,
    cluster_std = 0.6,
    random_state = 42
)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    x[:,0],
    x[:,1],
    cmap='virisble',
)
```

```python
model = HDBSCAN(min_cluster_size = 5)
```

```python
model.fit_predict(x)
```

```python
labels = model.labels_
```

```python
print(labels)
```

```python
plt.Figure(figsize=(8,6))

plt.scatter(
    x[:,0],
    x[:,1],
    c = labels
)
```
